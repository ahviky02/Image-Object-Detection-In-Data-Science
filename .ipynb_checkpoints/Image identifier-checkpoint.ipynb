```python
#this function is to resize image and convert image to array for preprocessing 

def load_and_preprocess_image(file_path, target_size=(224, 224)):
    try:
        image_array = cv2.imread(file_path)
        if image_array is not None:
            # Resize the image to the target size
            image_array = cv2.resize(image_array, target_size)
            return image_array
        else:
            print(f"Error: Unable to load image from {file_path}")
            return None
    except Exception as e:
        print(f"Error processing image {file_path}: {str(e)}")
        return None
```

```python
#this is get data from folder and convert to image to aaray 
directory = os.listdir('./raw-img')
data = {}

for item in directory:
    path = f"./raw-img/{item}"
    images = []  # Initialize a list to store images for each category
    for subdir, _, files in os.walk(path):
        for file in files:
            file_path = os.path.join(subdir, file)
            image = load_and_preprocess_image(file_path)
            if image is not None:
                images.append(image)
    data[item] = images

# Create a DataFrame from the dictionary with NaN for missing values
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
from tensorflow.keras.models import Sequential
from tensorflow import keras 
from keras import layers
from tensorflow.keras.layers import Activation, Dropout, Flatten, Dense 
from tensorflow.keras.layers import Conv2D, MaxPooling2D
```

```python
model = tf.keras.models.Sequential([ 
layers.Conv2D(32, (3, 3), activation='relu', input_shape=(200, 200, 3)), 
layers.MaxPooling2D(2, 2), 
layers.Conv2D(64, (3, 3), activation='relu'), 
layers.MaxPooling2D(2, 2), 
layers.Conv2D(64, (3, 3), activation='relu'), 
layers.MaxPooling2D(2, 2), 
layers.Conv2D(64, (3, 3), activation='relu'), 
layers.MaxPooling2D(2, 2), 

layers.Flatten(), 
layers.Dense(512, activation='relu'), 
layers.BatchNormalization(), 
layers.Dense(512, activation='relu'), 
layers.Dropout(0.1), 
layers.BatchNormalization(), 
layers.Dense(512, activation='relu'), 
layers.Dropout(0.2), 
layers.BatchNormalization(), 
layers.Dense(1, activation='sigmoid') 
]) 
```

```python
model.summary()
```

```python
model.compile( 
    loss='binary_crossentropy', 
    optimizer='adam', 
    metrics=['accuracy'] 
) 
```

```python
history = model.fit(train_datagen, 
    epochs=5, 
    validation_data=test_datagen) 
```
